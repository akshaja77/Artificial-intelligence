```python
# @title Default title text
#Basic vacuum cleaner

agent_table = {
    ('Clean', 'A'): 'MoveRight',
    ('Clean', 'B'): 'MoveLeft',
    ('Dirty', 'A'): 'Suck',
    ('Dirty', 'B'): 'Suck',
}

class VacuumCleaner:
    def __init__(self, location='A', room_statuses={'A': 'Dirty', 'B': 'Dirty'}):
        self.location = location
        self.room_statuses = room_statuses

    def percept(self):
        return self.room_statuses[self.location], self.location

    def act(self, action):
        if action == 'MoveRight':
            self.location = 'B'
        elif action == 'MoveLeft':
            self.location = 'A'
        elif action == 'Suck':
            self.room_statuses[self.location] = 'Clean'

    def get_cleaning_percentage(self):
        total_rooms = len(self.room_statuses)
        clean_rooms = sum(1 for status in self.room_statuses.values() if status == 'Clean')
        if total_rooms == 0:
            return 0.0
        return (clean_rooms / total_rooms) * 100

    def get_overall_cleaning_status(self):
        if self.get_cleaning_percentage() == 100:
            return "The entire floor is CLEAN!"
        else:
            return "The floor is not yet clean."

def table_driven_agent(percept):
    return agent_table.get(percept, 'NoOp')

if __name__ == "__main__":
    vacuum = VacuumCleaner(location='A', room_statuses={'A': 'Dirty', 'B': 'Dirty'})
    for step in range(5):
        percept = vacuum.percept()
        action = table_driven_agent(percept)
        print(f"Step {step+1}: Percept {percept}, Action: {action}")
        if action != 'NoOp':
            vacuum.act(action)

        print(f"Current Room Status: {vacuum.room_statuses}")
        print(f"Floor Cleaned: {vacuum.get_cleaning_percentage():.2f}%")
        print(f"Overall Status: {vacuum.get_overall_cleaning_status()}\n")
```

```python
#Advanced vacuum cleaner
# 0 = Clean, 1 = Dirt

rooms = {
    "Kitchen":     [1, 1, 1, 1],
    "Hall":        [1, 1, 1, 1],
    "Bedroom 1":   [1, 1, 1, 1],
    "Bedroom 2":   [1, 1, 1, 1],
    "Store Room":  [1, 1, 1, 1],
    "Terrace":     [1, 1, 1, 1]
}

performance = 0
total_cells = 0

# Count total cells
for room in rooms.values():
    total_cells += len(room)

positions = {
    0: "LEFT",
    1: "CENTER-LEFT",
    2: "CENTER-RIGHT",
    3: "RIGHT"
}

def vacuum_agent(room_name, room):
    global performance
    print(f"\nEntering {room_name}")

    for i in range(len(room)):
        print(f" Moving to {positions[i]} position")

        if room[i] == 1:
            print("  Dirt found → SUCK")
            room[i] = 0
            performance += 1

def room_status(room):
    clean = room.count(0)
    dirty = room.count(1)
    neat_percent = (clean / len(room)) * 100
    dirty_percent = (dirty / len(room)) * 100
    return neat_percent, dirty_percent


# -------- INITIAL STATUS --------
print("INITIAL STATUS\n")

for name, room in rooms.items():
    neat, dirty = room_status(room)
    print(f"{name}: {room} | Clean: {neat:.0f}% | Dirty: {dirty:.0f}%")

# -------- CLEANING PROCESS --------
print("\nCLEANING STARTED")

for name, room in rooms.items():
    vacuum_agent(name, room)

# -------- FINAL STATUS --------
print("\nFINAL STATUS\n")

for name, room in rooms.items():
    neat, dirty = room_status(room)
    print(f"{name}: {room} | Clean: {neat:.0f}% | Dirty: {dirty:.0f}%")

# -------- PERFORMANCE --------
performance_percent = (performance / total_cells) * 100

print(f"\nPerformance Measure: {performance_percent:.2f}% cleaned")
```

```python
#Basic BFS
from collections import deque

# BFS algorithm
def bfs(graph, root):
    visited = set()
    queue = deque([root])
    visited.add(root)

    parent = {root: None}   # to store BFS tree

    print("\nBFS Traversal:")
    while queue:
        vertex = queue.popleft()
        print(vertex, end=" ")

        for neighbour in graph[vertex]:
            if neighbour not in visited:
                visited.add(neighbour)
                parent[neighbour] = vertex
                queue.append(neighbour)

    return parent


# Function to print BFS tree diagram
def print_bfs_tree(parent):
    print("\n\nBFS Tree (Parent → Child):")
    for node in parent:
        if parent[node] is not None:
            print(f"{parent[node]} → {node}")


# ----------- USER INPUT -------------
n = int(input("Enter number of nodes: "))
nodes = list(map(int, input("Enter node values: ").split()))

graph = {node: [] for node in nodes}

m = int(input("Enter number of edges: "))
print("Enter edges (u v):")
for _ in range(m):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)   # remove this line if graph is directed

start = int(input("Enter starting node: "))

parent = bfs(graph, start)
print_bfs_tree(parent)
```

```python
#Advanced BFS
from collections import deque

# Input section
n = int(input("Enter the number of nodes: "))
v = input("Enter the node names separated by space: ").split()

g = {x: [] for x in v}

m = int(input("Enter the number of edges: "))

print("Enter the edges (node1 node2):")
for _ in range(m):
    a, b = input().split()
    g[a].append(b)
    g[b].append(a)

start = input("Enter the starting node: ")

# BFS Initialization
q = deque([start])
visited = set()
parent = {start: None}

step = 1

def get_path(node):
    path = []
    while node is not None:
        path.append(node)
        node = parent[node]
    return path[::-1]

# BFS Process
while q:
    x = q.popleft()
    visited.add(x)

    unvisited = set(v) - visited

    print(f"\nStep {step}")
    print("Source node :", x)
    print("Adjacent    :", g[x])
    print("Visited     :", visited)
    print("Unvisited   :", unvisited)
    print("Path        :", " -> ".join(get_path(x)))

    for y in g[x]:
        if y not in visited and y not in q:
            q.append(y)
            parent[y] = x

    step += 1

# Final Output
print("\nOptimal Solution (Shortest Paths from Source):")
for node in v:
    if node in parent:
        print(f"{start} -> {node} : {' -> '.join(get_path(node))}")
    else:
        print(f"{start} -> {node} : No path")
```

```python
#Uniform cost search
import heapq

def uniform_cost_search(graph, start, goal):
    frontier = []
    heapq.heappush(frontier, (0, start, [start]))
    explored = set()

    while frontier:
        cost, node, path = heapq.heappop(frontier)

        if node == goal:
            return path, cost

        if node not in explored:
            explored.add(node)

            for neighbour, edge_cost in graph[node]:
                if neighbour not in explored:
                    heapq.heappush(
                        frontier,
                        (cost + edge_cost, neighbour, path + [neighbour])
                    )

    return None, float("inf")


# -------- Main Program --------

graph = {}

n = int(input("Enter number of nodes: "))
for _ in range(n):
    node = input("Enter node name: ")
    graph[node] = []

e = int(input("Enter number of edges: "))
print("Enter edges (source destination cost):")

for _ in range(e):
    src, dest, cost = input().split()
    cost = int(cost)
    graph[src].append((dest, cost))
    # Uncomment below if graph is undirected
    # graph[dest].append((src, cost))

start = input("Enter start node: ")
goal = input("Enter goal node: ")

path, cost = uniform_cost_search(graph, start, goal)

if path:
    print("Path found:", " -> ".join(path))
    print("Total cost:", cost)
else:
    print("No path found")
```

```python
#Monkey & Banana problem
from collections import deque

# Possible places
PLACES = ["door", "window", "center"]

# State format:
# (monkey_position, box_position, on_what, has_banana)

INITIAL_STATE = ("door", "window", "floor", False)

# Goal: monkey has banana
GOAL_STATE = lambda s: s[3] == True


# -------------------- GET ACTIONS --------------------
def get_actions(state):
    actions = []
    m, b, on, has = state

    # Monkey can walk if on floor
    if on == "floor":
        for p in PLACES:
            if p != m:
                actions.append(("walk", p))

    # Monkey can push box if monkey and box at same place and on floor
    if m == b and on == "floor":
        for p in PLACES:
            if p != m:
                actions.append(("push", p))

    # Monkey can climb box
    if m == b and on == "floor":
        actions.append(("climb", None))

    # Monkey can grab banana
    if m == "center" and b == "center" and on == "box" and not has:
        actions.append(("grab", None))

    return actions


# -------------------- APPLY ACTION --------------------
def apply_action(state, action):
    m, b, on, has = state
    name, target = action

    if name == "walk":
        return (target, b, "floor", has)

    elif name == "push":
        return (target, target, "floor", has)

    elif name == "climb":
        return (m, b, "box", has)

    elif name == "grab":
        return (m, b, on, True)

    return state


# -------------------- BFS SOLVE --------------------
def solve():
    queue = deque()
    queue.append((INITIAL_STATE, []))
    visited = set()

    while queue:
        state, plan = queue.popleft()

        if state in visited:
            continue

        visited.add(state)

        if GOAL_STATE(state):
            return plan, state

        for action in get_actions(state):
            new_state = apply_action(state, action)
            queue.append((new_state, plan + [action]))

    return None, None


# -------------------- MAIN --------------------
if __name__ == "__main__":
    plan, final_state = solve()

    print("Initial State:", INITIAL_STATE)
    print("\nPlan to get banana:\n")

    for step in plan:
        print(step)

    print("\nFinal State:", final_state)
```

```python
#8-Puzzle
import heapq

goal = ((1,2,3),
        (8,0,4),
        (7,6,5))

# Manhattan Distance
def h(state):
    d = 0
    for i in range(3):
        for j in range(3):
            val = state[i][j]
            if val != 0:
                for x in range(3):
                    for y in range(3):
                        if goal[x][y] == val:
                            d += abs(x-i) + abs(y-j)
    return d

def find_blank(s):
    for i in range(3):
        for j in range(3):
            if s[i][j] == 0:
                return i, j

def astar(start):
    pq = []
    heapq.heappush(pq, (h(start), 0, start, "START"))
    visited = set()

    while pq:
        f, g, state, move = heapq.heappop(pq)

        print("Move:", move)
        print("g =", g)
        print("h =", h(state))
        print("f =", f)
        for row in state:
            print(row)
        print("--------------")

        if state == goal:
            print("GOAL REACHED")
            return

        visited.add(state)

        x, y = find_blank(state)
        moves = [(-1,0,"UP"), (1,0,"DOWN"),
                 (0,-1,"LEFT"), (0,1,"RIGHT")]

        for dx, dy, name in moves:
            nx, ny = x+dx, y+dy
            if 0<=nx<3 and 0<=ny<3:
                temp = [list(r) for r in state]
                temp[x][y], temp[nx][ny] = temp[nx][ny], temp[x][y]
                new_state = tuple(tuple(r) for r in temp)

                if new_state not in visited:
                    heapq.heappush(pq,
                        (g+1+h(new_state), g+1, new_state, name))

start = ((2,8,3),
         (1,6,4),
         (7,0,5))

astar(start)
```
